```python
import tweepy as tw
import csv
import pandas as pd
```

```python
#here you need to implement your developer keys
consumer_key="xxxxxxxxxxxxxxxxxxxxxxx"
consumer_secret="xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"
access_token="xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"
access_token_secret="xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"
```

```python
auth=tw.OAuthHandler(consumer_key,consumer_secret)
auth.set_access_token(access_token,access_token_secret)
api=tw.API(auth)
```

```python
quantity=500
tweets=[]
likes=[]
time=[]

#here we are mentioning the spreadsheet's name
file_name='Crypto_Currencies'

#here we are opening the spreadsheet where we are gonna store the scraped data
with open('%s.csv' %(file_name), 'w', encoding='utf-8') as file:
    w=csv.writer(file)
    
    #creating header row of spreadsheet and naming a particular column
    w.writerow(['tweets','likes','time'])
    
    #for each tweet matching out searching critera(hashtags), write relevant information to the datasheet
    #as a result we are implementing a cursor that will scrape data acording to our needs and search only the relevant topics 
    #and filter out retweets and search only in the language that we want to, here I am only searching tweets that are available
    #only in english language.
    #you can modify the search according to your needs by modifying the cursor, in api search you can change some syntax 
    #to scarpe data acording to your needs.
    for i in tw.Cursor(api.search, q='Crypto Currencies -filter:retweets', lang='en', tweet_mode='extended').items(quantity):
        tweets.append(i.full_text)
        likes.append(i.favorite_count)
        time.append(i.created_at)
        w.writerow([i.full_text.replace('\n','  ').encode('utf-8'), i.favorite_count, i.created_at])
```

```python
df=pd.DataFrame({'tweets':tweets,'likes':likes,'time':time})
```
